# <a name="HW4Q3">Question 3</a>

------------------------------------------------------------------------

The variance of a continuous random variable $X$, with pdf $f(x)$ , is defined as

$$\mbox{Var} (X) = \int_{-\infty}^{\infty} \bigg( (x-\mu)^2 \cdot f(x) \bigg) \, dx ,$$ where $\mu=E(X) = \int_{-\infty}^{\infty} x \cdot f(x) \, dx$ is the expected value of $X$.

Prove that $$ \int_{-\infty}^{\infty} \bigg( (x-\mu)^2 \cdot f(x) \bigg) \, dx = E(X^2) - (E(X))^2.$$

## Solution to Question 3

------------------------------------------------------------------------


$$\begin{align}
\int_{-\infty}^{\infty} \bigg( (x-\mu)^2 \cdot f(x) \bigg) \, dx &= \int_{-\infty}^{\infty} \bigg( (x^2-2\mu x + \mu^2) \cdot f(x) \bigg) \, dx & \quad \mbox{Distributing the product $(x-\mu)^2$} \\
&= \int_{-\infty}^{\infty} \left( x^2f(x) - 2\mu x f(x) + \mu^2 f(x) \right) \, dx & \quad \mbox{Justify this step in Explanation 1} \\
&= \int_{-\infty}^{\infty} x^2f(x) \, dx - \int_{-\infty}^{\infty} 2\mu x f(x) \, dx + \int_{-\infty}^{\infty} \mu^2 f(x) \, dx & \quad \mbox{Justify this step in Explanation 2} \\
&= \left( \int_{-\infty}^{\infty} x^2f(x) \, dx \right) - \left( 2\mu \int_{-\infty}^{\infty} x f(x) \, dx \right) + \left( \mu^2 \int_{-\infty}^{\infty} f(x) \, dx \right) & \quad \mbox{Justify this step in Explanation 3} \\
&= \left( \int_{-\infty}^{\infty} x^2f(x) \, dx \right) - \left( 2\mu \int_{-\infty}^{\infty} x f(x) \, dx \right) + \mu^2 & \quad \mbox{Justify this step in Explanation 4} \\
&= \left( \int_{-\infty}^{\infty} x^2f(x) \, dx \right) - 2\mu^2 + \mu^2 & \quad \mbox{Justify this step in Explanation 5} \\
&= E(X^2) - 2\mu^2 + \mu^2 & \quad \mbox{Justify this step in Explanation 6} \\
&= E(X^2) - \big( E(X) \big)^2 & \quad \mbox{Justify this step in Explanation 7}
\end{align}$$


### Explanations for Solution to Question 3

Pulling out the constant multiplier(mu) from the integrals that contain this constant.

**Explanation 5:**
