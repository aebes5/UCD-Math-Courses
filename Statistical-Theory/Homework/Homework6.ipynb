----

Complete the partially completed code in the first code cell to create a sampling distribution for the proportion using 10,000 random samples each size $n=60$. Then run the second code cell (without any edits needed)
to plot the sampling distribution with a vertical line through $\hat{p}=0.10$.

<br>

Based on the histogram you generated, **how many** samples have a sample proportion less than or equal to $0.06$?

<br>

Enter your count here: 147


```r
set.seed(mmdd)  # be sure you already defined mmdd in previous code cell above

samp.prop2b <- numeric(10000) # creates an empty vector to store results from 10,000 samples

# a for loop that generates 10,000 random samples

# Student To Do: Replace each ?? in the code below
for (i in 1:10000)
{
  temp.samp <- sample(c("Negative", "Positive"),  # pick random sample
                      size = 60,
                      replace = TRUE,
                      prob = c(0.85, 0.15))
  samp.prop2b[i] <- sum(temp.samp == "Positive") / 60 # calculate sample proportion
}
```

```r
# nothing to edit in this code cell
hist(samp.prop2b,  # plot stored sample proportions
     breaks = 19,  # use 8 breaks
     xlab = "Sample Proportion, P-Hat",  # x-axis label
     main = "Sampling Distribution of Proportion with Positive Test, n=50",  # main label
     labels = TRUE,  # add count labels above each bar
     xaxt='n',  # disable default x-axis
     yaxt='n')  # disable default y-axis

axis(1, at=seq(0, 0.4, 0.04), pos=0)  # customize ticks on x-axis
axis(2, at=seq(0, 4000, 500), pos=0)  # customize ticks on y-axis

abline(v = 0.10, col = "red", lwd = 2, lty = 2)  # draw vertical line at p-hat=0.10
```

# <a name="HW6q3">Question 3</a>

Web servers typically have security algorithms that detect and flag "abnormal" connections from suspicious IP addresses, which can indicate possible hackers. Data from the article "Exact Inferences for a Gamma Distribution" (*J. Quality Technol.* 2014: 140–149) suggests that, for
one particular server receiving abnormal connections from one specific IP address, the time $X$ in hours between two attempted connections can be modeled using a gamma distribution

$$X \sim \mbox{Gamma}(2, 0.4).$$

- The <font color="dodgerblue">shape parameter $r = 2$</font> means $X$ is the amount of <font color="dodgerblue">time it takes for two attempted connections</font>.
- The <font color="tomato">rate parameter $\lambda = \frac{1}{\mu} = 0.4$ </font> tells us that <font color="tomato">on average there are $0.4$ attempted connections per hour</font>.
- Equivalently $\mu = \frac{1}{\lambda} = 2.5$ tells us on average there is 1 attempted connection every 2.5 hours.
- The command `rgamma(n, r, lambda)` is used to randomly pick a sample of $n$ values from $X \sim \mbox{Gamma}(r, \lambda)$.



See more information about Gamma Distributions in the [Appendix of 08 Common Continuous Distributions](https://aspiegler.github.io/Statistical-Theory/08-Common-Continuous-Distributions.html#sec-append-gamma).

Complete the partially completed code cell below to create an approximate sampling distribution for the **median** weight time $X$ by generating 10,000 random samples each size $n=60$. After storing the distribution of sample medians to the vector `samp.median`:

- In the second code cell, compute the mean of your sampling distribution for the median.
- In the third code cell, compute the standard error of your sampling distribution for the median.

<br>

### Solution to Question 3a

---

Complete the code cells below and enter your answers in the bullets below. Round answers to four decimal places if needed.

<br>
<br>

- Mean of sampling distribution for the median: 4.2215
- Standard error of sampling distribution for the median: 0.5115


```r
set.seed(mmdd)  # be sure you already defined mmdd in previous code cell above

# creates an empty vector to store results
samp.median <- numeric(10000)

# a for loop that generates 10,000 random samples
# each size n=60, and calculates the sample median
for (i in 1:10000)
{
##################################
# Replace both ?? in the for loop
##################################
  sample.times <- rgamma(60, 2, 0.4)  # Step 1: Pick random sample
  samp.median[i] <- median(sample.times)  # Step 2: Calculate sample statistic
}
```

```r
# compute the mean of the sampling distribution
mean(samp.median)
```

```r
# compute the standard error of the sampling distribution
sd(samp.median)
```

```r
set.seed(mmdd)  # be sure you already defined mmdd in previous code cell above

# creates an empty vector to store results
samp.var <- numeric(10000)

# a for loop that generates 10,000 random samples
# each size n=60, and calculates the sample variance
for (i in 1:10000)
{
##################################
# Replace both ?? in the for loop
##################################
  sample.times <- rgamma(60, 2, 0.4)  # Step 1: Pick random sample
  samp.var[i] <- var(sample.times)  # Step 2: Calculate sample statistic
}
```
