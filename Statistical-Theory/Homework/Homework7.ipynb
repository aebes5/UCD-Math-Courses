```r
# Use theta for the Greek letter in the code below
# Hint: sqrt(x) is used to for square root of x
# Hint: exp(-1 * theta * sqrt(x) ) for the exponential term

# the sample is stored in the vector x
x <- c(3, 5, 8, 10)

# enter the likelihood function
like <- function(theta){
  pdf <- (theta / (2 * sqrt(x))  * exp(-1 * theta * sqrt(x)))  # enter a formula for the pdf
  prod(pdf)  # take the product of pdf's to get L(theta)
}
```

The for loop in the code cell below generates a distribution of MLE's for $p$ based on 10,000 random samples each size $n=15$ picked from $X \sim \mbox{Binom}(40, p)$ where $p$ has been stored as `true.p` in a previous code cell in [Question 5](#HW7q5).


<br>

### Solution to Question 5d

---

Complete the partially completed code cell below and run it to generate and plot a distribution of MLE's.


- There are three `??` that you need to replace in the for loop before running.
- You do not need to provide any other supporting work.

<br>  
<br>

```r
binom.mle <- numeric(10000)  # empty numeric vector where each estimate is stored

#####################################################################
# Student To Do: Replace each of the three ??'s with appropriate code
#####################################################################
for (i in 1:10000)
{
  temp.x <- rbinom(15, 40, true.p)  # pick a random sample size n=15 from X ~ Binom(40, p)
  like.binom <- function(p){
    pmf.binom <- dbinom(temp.x, 40, p)  # define pmf
    prod(pmf.binom)  # take prod of pmf's to get L(p)
}
  binom.mle[i] <- optimize( like.binom, c(0,1), maximum = TRUE )$maximum  # find value of MLE for p
}

# display results as histogram
# nothing further to edit
hist(binom.mle,
     breaks = 20,
     xlab = "p_MLE",
     main = "Dist. of MLE's for Binomial Dist")
abline(v = true.p, col = "blue", lwd = 2)  # plot at actual value of lambda
```
