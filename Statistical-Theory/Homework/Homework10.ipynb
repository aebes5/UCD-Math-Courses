In the 1970s, research<sup>1</sup> by Alfred Sommer on vitamin A and night blindness suggested that vitamin A also reduced childhood death rates. Children were randomly assigned to two groups, one group took a vitamin A pill every day, and children in the other group were given a **placebo** (a pill that had no vitamin A). Below is a two-way table displaying the results of Sommer's research.


|       | Vitamin A | No Vitamin A | Total  |
|-------|-----------|--------------|--------|
| Died  | 101       | 130          | 231    |
| Lived | 12,890    | 12,079       | 24,969 |
| Total | 12,991    | 12,209       | 25,200 |


<br>

<font size=2>1. [www.laskerfoundation.org/rprimers/sommer/saving1.htm](www.laskerfoundation.org/rprimers/sommer/saving1.htm) and *Introduction to the Practice of Statistics*, 4th ed, by David Moore, George. McCabe</font>


## <a name="HW10q1a">Question 1a</a>

---

Find a 95% confidence interval for the difference in the proportion of children that died between the two groups (received vitamin A and did not receive vitamin A) using the Wald confidence interval for the difference between two proportions.

### Solution to Question 1a

---

<br>

**Complete the R code cell below to compute the cutoffs for a 95% Wald Confidence Interval.**

<br>  


```r
# Calculate the sample proportion for each separately
treat.p <- 101 / 12991  # proportion in treatment (vitamin A) group that died
control.p <- 130 / 12209  # proportion in control (no vitamin A) group that died

# Approximate the SE using Wald Conf Int
wald.se <- sqrt((treat.p * (1 - treat.p) / 12991) + control.p * (1 - control.p) / 12209)

# Compute the z-alpha-over-2 value for 95% Conf Int
z.star <- qnorm(0.975, 0, 1)

# Compute the Margin of Error
moe <- z.star * wald.se

# Compute upper and lower cutoffs for Conf Int
lower.q1 <- treat.p - control.p - moe
upper.q1 <- treat.p - control.p + moe

# Print answers to the screen
lower.q1
upper.q1
```
