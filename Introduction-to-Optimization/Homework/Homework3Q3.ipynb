```python
#PROBLEM 3

import pyomo.environ as pyo

model = pyo.ConcreteModel()

#decision variables
model.x1 = pyo.Var(domain=pyo.NonNegativeReals, bounds = (50, None))
model.x2 = pyo.Var(domain=pyo.NonNegativeReals)
model.x3 = pyo.Var(domain=pyo.NonNegativeReals)
model.x4 = pyo.Var(domain=pyo.NonNegativeReals)
model.y1 = pyo.Var(domain=pyo.NonNegativeReals)
model.y2 = pyo.Var(domain=pyo.NonNegativeReals)
model.y3 = pyo.Var(domain=pyo.NonNegativeReals)
model.y4 = pyo.Var(domain=pyo.NonNegativeReals)

#objective function
def obj_rule(model):
    return 5*model.x1 + 8*model.x2 + 2*model.y1 + 4*model.x3 + 2*model.y2 + 7*model.x4 + 2*model.y3 - 6*model.y4

model.obj = pyo.Objective(rule=obj_rule, sense=pyo.minimize)

#constraints
def enough_product_one(model):
    return model.x1 >= 50

model.enough_product_one_constraint = pyo.Constraint(rule=enough_product_one)

def enough_product_two(model):
    return (model.x1 + model.x2) >= 115

model.enough_product_two_constraint = pyo.Constraint(rule=enough_product_two)

def enough_product_three(model):
    return ((model.x1 + model.x2 + model.x3) >= 215)

model.enough_product_three_constraint = pyo.Constraint(rule=enough_product_three)

def enough_product_four(model):
    return (model.x1 + model.x2 + model.x3 + model.x4 == 285)

model.enough_product_four_constraint = pyo.Constraint(rule=enough_product_four)

def left_over_one(model):
    return (model.y1 == model.x1 - 50)

model.left_over_one_constraint = pyo.Constraint(rule=left_over_one)

def left_over_two(model):
    return (model.y2 == model.x1 + model.x2 - 115)

model.left_over_two_constraint = pyo.Constraint(rule=left_over_two)

def left_over_three(model):
    return (model.y3 == model.x1 + model.x2 + model.x3 - 215)

model.left_over_three_constraint = pyo.Constraint(rule=left_over_three)

def left_over_four(model):
    return (model.y4 == model.x1 + model.x2 + model.x3 + model.x4 - 285)

model.left_over_four_constraint = pyo.Constraint(rule=left_over_four)

result = pyo.SolverFactory("gurobi").solve(model, tee=True)



# Solution
print(f'The solver returned a status of: {result.solver.termination_condition}')

# Solution
if result.solver.termination_condition == pyo.TerminationCondition.optimal:
    print(pyo.TerminationCondition.optimal)
    print(f"Optimal value: {pyo.value(model.obj)}")
    print("Optimal solution:")
    print(f"  x1: {pyo.value(model.x1)}")
    print(f"  x2: {pyo.value(model.x2)}")    
    print(f"  x3: {pyo.value(model.x3)}")    
    print(f"  x4: {pyo.value(model.x4)}")   
    print(f"  y1: {pyo.value(model.y1)}")
    print(f"  y2: {pyo.value(model.y2)}")    
    print(f"  y3: {pyo.value(model.y3)}")    
    print(f"  y4: {pyo.value(model.y4)}")    
    
print('We should do:')
print(f'{pyo.value(model.x1)} products produced in month 1')
print(f'{pyo.value(model.x2)} products produced in month 2')
print(f'{pyo.value(model.x3)} products produced in month 3')
print(f'{pyo.value(model.x4)} products produced in month 4')
print(f'{pyo.value(model.y1)} products left over in month 1')
print(f'{pyo.value(model.y2)} products left over in month 2')
print(f'{pyo.value(model.y3)} products left over in month 3')
print(f'{pyo.value(model.y4)} products left over in month 4')
print(f"for a cost of ${pyo.value(model.obj)}")
```
