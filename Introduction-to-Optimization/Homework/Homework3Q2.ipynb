```python
#PROBLEM 2

import pyomo.environ as pyo

model = pyo.ConcreteModel()

#decision variables
model.A = pyo.Var(domain=pyo.NonNegativeReals, bounds = (0,1000))
model.B = pyo.Var(domain=pyo.NonNegativeReals, bounds = (0,1000))
model.C = pyo.Var(domain=pyo.NonNegativeReals, bounds = (0,1000))
model.D = pyo.Var(domain=pyo.NonNegativeReals)

#objective function
def obj_rule(model):
    return 0.424*model.A + 0.42*model.B + 0.43*model.C - 8.005*model.D

model.obj = pyo.Objective(rule=obj_rule, sense=pyo.maximize)


#constraints
def silicon(model):
    return ((0.45*model.A + 0.5*model.B + 0.4*model.C + 100*model.D)/1000) >= 0.45

model.silicon_constraint = pyo.Constraint(rule=silicon)

def manganese_lower(model):
    return ((4*model.A + model.B + 0.6*model.C)/1000) >= 3.25

model.manganese_constraint_lower = pyo.Constraint(rule=manganese_lower)

def manganese_upper(model):
    return ((4*model.A + model.B + 0.6*model.C)/1000) <= 5.5

model.manganese_constraint_upper = pyo.Constraint(rule=manganese_upper)

def exact(model):
    return model.A + model.B + model.C + model.D == 1000

model.exact_constraint = pyo.Constraint(rule = exact)

result = pyo.SolverFactory("gurobi").solve(model, tee=True)



# Solution
print(f'The solver returned a status of: {result.solver.termination_condition}')

# Solution
if result.solver.termination_condition == pyo.TerminationCondition.optimal:
    print(pyo.TerminationCondition.optimal)
    print(f"Optimal value: {pyo.value(model.obj)}")
    print("Optimal solution:")
    print(f"  A: {pyo.value(model.A)}")
    print(f"  B: {pyo.value(model.B)}")    
    print(f"  C: {pyo.value(model.C)}")    
    print(f"  D: {pyo.value(model.D)}")    
    
print('We should do:')
print(f'{pyo.value(model.A)} pounds of pig iron A')
print(f'{pyo.value(model.B)} pounds of pig iron B')
print(f'{pyo.value(model.C)} pounds of pig iron C')
print(f'{pyo.value(model.D)} pounds of pure manganese')
print(f"for a profit of ${pyo.value(model.obj)}")
```
