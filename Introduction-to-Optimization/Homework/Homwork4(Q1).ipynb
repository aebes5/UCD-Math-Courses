```python
#PROBLEM 1

import pyomo.environ as pyo

model = pyo.ConcreteModel()

#decision variables
model.x1 = pyo.Var(domain=pyo.NonNegativeReals)
model.x2 = pyo.Var(domain=pyo.NonNegativeReals)
model.x3 = pyo.Var(domain=pyo.NonNegativeReals)
model.x4 = pyo.Var(domain=pyo.NonNegativeReals)
model.y1 = pyo.Var(domain=pyo.NonNegativeReals)
model.y2 = pyo.Var(domain=pyo.NonNegativeReals)
model.y3 = pyo.Var(domain=pyo.NonNegativeReals)
model.y4 = pyo.Var(domain=pyo.NonNegativeReals)


#objective function
def obj_rule(model):
    return 12 * 32 * (model.x1 + model.x2 + model.x3 + model.x4 + model.y1 + model.y2 + model.y3 + model.y4) + 6 * 44 * (model.x1 + model.x2 + model.x3 + model.x4)

model.obj = pyo.Objective(rule=obj_rule, sense=pyo.minimize)

#constraints
def nurses_12am(model):
    return  model.x1 + model.x4 + model.y1 + model.y1 + model.y3 + model.y4 >= 12

model.nurses_12am_const = pyo.Constraint(rule=nurses_12am)

def nurses_6am(model):
    return  model.x1 + model.x2 + model.y1 + model.y1 + model.y2 + model.y4 >= 8

model.nurses_6am_const = pyo.Constraint(rule=nurses_6am)

def nurses_12pm(model):
    return  model.x2 + model.x3 + model.y1 + model.y1 + model.y2 + model.y3 >= 6

model.nurses_12pm_const = pyo.Constraint(rule=nurses_12pm)

def nurses_6pm(model):
    return  model.x3 + model.x4 + model.y1 + model.y2 + model.y3 + model.y4 >= 15

model.nurses_6pm_const = pyo.Constraint(rule=nurses_6pm)


result = pyo.SolverFactory("gurobi").solve(model, tee=True)



# Solution
print(f'The solver returned a status of: {result.solver.termination_condition}')

# Solution
if result.solver.termination_condition == pyo.TerminationCondition.optimal:
    print(pyo.TerminationCondition.optimal)
    print(f"Optimal value: {pyo.value(model.obj)}")
    print("Optimal solution:")
    print(f"  x1: {pyo.value(model.x1)}")
    print(f"  x2: {pyo.value(model.x2)}")    
    print(f"  x3: {pyo.value(model.x3)}")    
    print(f"  x4: {pyo.value(model.x4)}")    
    print(f"  y1: {pyo.value(model.y1)}")
    print(f"  y2: {pyo.value(model.y2)}")    
    print(f"  y3: {pyo.value(model.y3)}")    
    print(f"  y4: {pyo.value(model.y4)}")    
    
print('We should do:')
print(f'{pyo.value(model.x1)} nurses should start a 12 hour shift at 12 midnight')
print(f'{pyo.value(model.x2)} nurses should start a 12 hour shift at 6 am')
print(f'{pyo.value(model.x3)} nurses should start a 12 hour shift at 12 noon')
print(f'{pyo.value(model.x4)} nurses should start a 12 hour shift at 6 pm')
print(f'{pyo.value(model.y1)} nurses should start a 18 hour shift at 12 midnight')
print(f'{pyo.value(model.y2)} nurses should start a 18 hour shift at 6 am')
print(f'{pyo.value(model.y3)} nurses should start a 18 hour shift at 12 noon')
print(f'{pyo.value(model.y4)} nurses should start a 18 hour shift at 6 pm')
print(f"for a cost of ${pyo.value(model.obj)}")
```
