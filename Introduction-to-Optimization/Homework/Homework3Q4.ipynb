```python
#PROBLEM 4a

import pyomo.environ as pyo

model = pyo.ConcreteModel()

#decision variables
model.A = pyo.Var(domain=pyo.NonNegativeReals)
model.B = pyo.Var(domain=pyo.NonNegativeReals)
model.C = pyo.Var(domain=pyo.NonNegativeReals)
model.D = pyo.Var(domain=pyo.NonNegativeReals)
model.E = pyo.Var(domain=pyo.NonNegativeReals)
model.F = pyo.Var(domain=pyo.NonNegativeReals)
model.G = pyo.Var(domain=pyo.NonNegativeReals)


#objective function
def obj_rule(model):
    return 10*model.A + 11*model.B + 12*model.C + 13*model.D + 14*model.E + 12*model.F + 15*model.G

model.obj = pyo.Objective(rule=obj_rule, sense=pyo.minimize)

#constraints
def cane_sugar(model):
    return  .1*model.A + .1*model.B + .2*model.C + .3*model.D + .4*model.E + .2*model.F + .6*model.G == 52

model.cane_sugar_constraint = pyo.Constraint(rule=cane_sugar)

def corn_sugar(model):
    return  .3*model.A + .4*model.B + .4*model.C + .2*model.D + .6*model.E + .7*model.F + .1*model.G == 56

model.corn_sugar_constraint = pyo.Constraint(rule=corn_sugar)

def beet_sugar(model):
    return  .6*model.A + .5*model.B + .4*model.C + .5*model.D + .1*model.F + .3*model.G == 59

model.beet_sugar_constriant = pyo.Constraint(rule=beet_sugar)


result = pyo.SolverFactory("gurobi").solve(model, tee=True)



# Solution
print(f'The solver returned a status of: {result.solver.termination_condition}')

# Solution
if result.solver.termination_condition == pyo.TerminationCondition.optimal:
    print(pyo.TerminationCondition.optimal)
    print(f"Optimal value: {pyo.value(model.obj)}")
    print("Optimal solution:")
    print(f"  A: {pyo.value(model.A)}")
    print(f"  B: {pyo.value(model.B)}")    
    print(f"  C: {pyo.value(model.C)}")    
    print(f"  D: {pyo.value(model.D)}")    
    print(f"  E: {pyo.value(model.E)}")
    print(f"  F: {pyo.value(model.F)}")    
    print(f"  G: {pyo.value(model.G)}")    
    
print('We should do:')
print(f'{pyo.value(model.A)} tons of sugar A')
print(f'{pyo.value(model.B)} tons of sugar B')
print(f'{pyo.value(model.C)} tons of sugar C')
print(f'{pyo.value(model.D)} tons of sugar D')
print(f'{pyo.value(model.E)} tons of sugar E')
print(f'{pyo.value(model.F)} tons of sugar F')
print(f'{pyo.value(model.G)} tons of sugar G')
print(f"for a cost of ${pyo.value(model.obj)}")
```
